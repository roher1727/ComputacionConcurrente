```python
from threading import Thread, Lock, Condition
import time
import random
import queue
```

```python
class Consumer(Thread):
    
    def __init__(self, condition):
        Thread.__init__(self)
        self.mutex = condition
    
    
    def run(self):
        global queue
        while True:
            mutex.acquire()
            if not queue:
                print ("Nada en la cola, consumidor esperando")
                mutex.wait()
                print ("Productor agrego algo a la cola y notifico al consumidor")
            num = queue.pop(0)
            print ("Consumido", num)
            mutex.release()
            time.sleep(random.random())
```

```python
class Producer(Thread):
    def __init__(self, condition):
        Thread.__init__(self)
        self.mutex = condition
    
    def run(self):
        nums = range(5)
        global queue
        while True:
            self.mutex.acquire()
            num = random.choice(nums)
            queue.append(num)
            print ("Producido", num)
            self.mutex.notify()
            self.mutex.release()
            time.sleep(random.random())
```

```python
queue = []
mutex = Condition()

Producer(mutex).start()
Producer(mutex).start()
Consumer(mutex).start()
```
